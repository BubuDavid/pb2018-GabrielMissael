```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline 
```

```python
n = 20
l = 0.1
s=1000
h = n/s
w = (9.81/l)**(1/2)
ini = 80
v = np.linspace(0, s, s+1)
t = np.linspace(0, s, s+1)
the = np.linspace(0, s, s+1)
v[0] = 0
the[0]= ini
```

```python
def f1(ang):
    return h*(-w**2*ang)

def f2(i):
    return h*v[i]

def armonico():
    for i in range(s-1):
        v[i+1] = v[i]+f1(the[i])
        the[i+1]= the[i]+f2(i)      
```

```python
armonico()
```

```python
plt.plot(t, the)
```

```python
plt.plot(t, v)
```

```python
plt.plot(v, the)
```
