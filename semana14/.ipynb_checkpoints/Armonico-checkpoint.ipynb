```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline 
```

```python
n = 25
#segundos de analizis
h = 0.001
#cambio de tiempo
l = 2
#longitud de la cuerda
ini = 20.0
#angulo inicial
s= n/h
#numero total de pasos
s=int(s)
w = (9.81/l)**(0.5)
#frecuencia angular
v = np.linspace(0, s, s+1)
t = np.linspace(0, s, s+1)
the = np.linspace(0, s, s+1)
v[0] = 0.0
the[0]= ini
```

```python
def f1(ang):
    return h*(-w**2*ang)

def f2(i):
    return h*v[i]

def armonico():
    for i in range(s):
        v[i+1] = v[i]+f1(the[i])
        the[i+1]= the[i]+f2(i)
        if(the[i+1]>ini):
            the[i+1]=ini
        if(the[i+1]<-ini):
            the[i+1]=-ini
```

```python
armonico()
```

```python
plt.plot(t, the)
plt.grid()
```

```python
plt.plot(t, v)
plt.grid()
```

```python
plt.plot(v, the)
plt.grid()
```
