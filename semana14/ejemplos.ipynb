```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline 
#para activar los graficos en el notebook 
```

```python
x = np.linspace(-10, 10, 100)
x2 = np.linspace(0, 5, 100)
y = x**2*np.cos(x)-3*x**3
y2 = x2**2*np.cos(x2)-3*x2**3
```

```python
plt.plot(x,y)
plt.plot(x2,y2)
```

```python
#OPcion 1 para definir funciones
fx = lambda x: x**2*np.cos(x)-3*x**3
```

```python
#print(fx(x2))
```

```python
plt.plot(x,fx(x))
plt.plot(x2, fx(x2))
```

```python
f2x = lambda x,y: x*y*np.cos(x/y)-3*x**3
```

```python
plt.plot(x, f2x(x,y))
plt.plot(x, f2x(y,x))
```

```python
#Segunda opcion para definir funciones
def f3x(x,y):
    z=x*y*np.cos(x/y)-3*x**3
    return z
```

```python
plt.plot(x, f3x(x,y))
```

```python
x3 = np.random.rand(10)
y3 = 2
plt.plot(x3, f3x(x3, y3))
```

```python
#b=0 por defecto a menos que especifiquemos otra cosa
def recta(a, m, b=0):
    return m*a+b
```

```python
x = np.linspace(-5, 5, 10)
plt.plot(x, recta(x, 3))
plt.plot(x, recta(x, 1, -3))
plt.plot(x, recta(x, -3, 6))
plt.plot(x, recta(m=9, b=1, a=x))
plt.grid()
```

```python
def recta2(a, m=1, b=0, label='x'):
    if(m<0):
        m=0
    x = m*a+b
    plt.plot(a, x, label=label)
    return
```

```python
x = np.linspace(-2, 2, 10)
recta2(x)
recta2(x, -2, 1, '-x+1')
plt.legend()
plt.grid()
```
