```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.constants as con

%matplotlib inline 
```

```python
puntos = np.loadtxt("IRCF.txt").T
puntos[3][42]=200
N = np.size(puntos[1])
```

```python
def grafica(b, a):
    
    f = plt.figure(figsize=(12.5,10.5))
    
    #f, axs = plt.subplots(2,1, figsize(15.5, 10.5), gridspec_kw = {'height_ratios':[4, 1]}, sharex=True)
    ceros = np.zeros(N)
    
    #ax1 = axs[1][0]
    ax1 = f.add_subplot(211)
    ax1.set_title('Fondo cosmico de microondas')
    ax1.scatter(puntos[0], puntos[1])
    for i in range (a):
        ax1.plot(puntos[0], b[i])
    ax1.grid()
    ax1.set_ylabel('Intensidad (MJy/sr)')
    ax1.set_xlabel('Frecuencia (1/cm)')
    
    #ax2 = axs[1][1]
    ax2 = f.add_subplot(212)
    ax2.errorbar(puntos[0], ceros, puntos[3]/1000)
    ax2.grid()
    ax2.set_xlabel('Frecuencia (1/cm)')
    ax2.set_ylabel('Incertidumbre (MJy/sr)')
    
    return
```

```python
def Int(v, T):
    h = con.h
    c = con.c
    k = con.k
    v *= (c*100)
    factor = (2*h*v**3/c**2)*1e26
    arg=(h*v)/(k*T)
    I = factor/(np.exp(arg)-1)
    return I/1e6
```

```python
def temp(t):
    y=(np.zeros(N))
    for i in range (N):
        y[i] = Int((puntos[0][i]),t)
    return y
```

```python
res = np.zeros((11, N))
count =1
for i in range (11):
    res[i]=temp(0.3*count)
    count+=1
grafica(res, 11)
```
