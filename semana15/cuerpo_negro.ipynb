```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.constants as con

%matplotlib inline 
```

```python
puntos = np.loadtxt("IRCF.txt").T
```

```python
plt.scatter(puntos[0], puntos[1])
plt.grid()
```

```python
def grafica(b):
    axs = plt.subplots(nrows=2, ncols=1, sharex=True)
    ceros = np.zeros(np.size(puntos[0]))
    
    ax1 = axs[1][0]
    ax1.scatter(puntos[0], puntos[1])
    ax1.plot(puntos[0], b)
    ax1.grid()
    
    ax2 = axs[1][1]
    ax2.errorbar(puntos[0], ceros, puntos[3]/1000)
    ax2.grid()

    return
```

```python
grafica(puntos[1])
```

```python
def Int(v, T):
    h = con.h
    c = con.c
    k = con.k
    v *= (c*100)
    factor = (h*v**3/c**2)/1e-26
    arg=(h*v)/(k*T)
    I = factor/(np.exp(arg)-1)
    return I/1e6
```

```python
def temp(t):
    for i in range (np.size(puntos[1])):
        y[i] = Int(puntos[1][i],t)
    return y
```

```python
aux=temp(20)
print(aux)
print(puntos[0])
grafica(aux)
```
